```python
import numpy as np
import pandas as pd
```

```python
# This Linear Regression is for only single feature (2 Column) Datasets.

class SimpleLR():
    def _init_(self):
        self.w = 0
        self.b = 0
    def fit(self,X,Y):
        num = 0
        den = 0
        for x, y in zip(X.values, Y.values):
            num = num + (x - X.mean())*(y-Y.mean())
            den = den + (x - X.mean())**2
        self.w =  num/den
        self.b = Y.mean() - self.w * X.mean()
        print(self.w)
        print(self.b)
    def predict(self,X):
        print(X * self.w + self.b)
    
```

```python
df = pd.read_csv("salary_data.csv")
```

```python
X=df.iloc[:,0]
Y=df.iloc[:,1]
```

```python
lr = SimpleLR()
```

```python
lr.fit(X,Y)
```
